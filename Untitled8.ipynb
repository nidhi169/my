```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
from selenium import webdriver 
driver = webdriver.Chrome(PATH)
driver.get("https://archive.ics.uci.edu/ml/index.php")
```

```python
dataset_click = driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b')
dataset_click.click()
```

```python
dataset_name = []
for i in driver.find_elements_by_xpath('//table[@cellpadding = "3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]'):
    dataset_name.append(i.text)
```
